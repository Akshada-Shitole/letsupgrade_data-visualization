```python
"""Question 1

plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)

"""
import numpy as np
import matplotlib.pyplot as plt

a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,label="blue bar",color='b')
plt.title("line PLOT")
plt.xlabel("a")
plt.ylabel("b")
plt.show()

```

```python
"""
Question 2
Plot a line plot showing the sales trend in company 1 and 2:
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2

"""

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2

plt.plot(days,sales_1,label="blue bar",color='b')

plt.plot(days,sales_2,label="green bar",color='r')

plt.title("LINE PLOT")
plt.xlabel("days")
plt.ylabel("sales")

plt.show()



```
