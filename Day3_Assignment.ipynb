```python
"""
Question 1
Plot using sns.relplot, line plot, box plot on fmri data set
"""

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from numpy.random import randn,randint,uniform,sample
%matplotlib inline


sns.get_dataset_names()
```

```python
sns.set(style='darkgrid')
```

```python
data=sns.load_dataset('fmri')
data.head()
```

```python
sns.relplot(x='timepoint',y='signal',data=data,kind='line',hue='subject',style='event')
```

```python
sns.lineplot(x='timepoint',y='signal',data=data,hue='region',style='event')
```

```python
sns.boxplot(x='region',y='signal',data=data,hue='event')
```
